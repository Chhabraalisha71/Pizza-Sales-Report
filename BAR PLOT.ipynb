```python
#to draw bar graph for an example we are taking harry potter parts and its popularity 
# so first we have to make x and y axis
import matplotlib.pyplot as plt

x=["Part1","Part2","Part3","Part4","Part5"]
y=[98,95,86,87,52]
colors=["Red","Blue","Green","Yellow","Pink"]
plt.bar(x,y,color=colors,edgecolor="black")

plt.xlabel("Parts of Harry Potter",fontsize=10)
plt.ylabel("Popularity",fontsize=10)
plt.title("Popularity of different parts of harrypotter",fontsize=20)
plt.show()


```
